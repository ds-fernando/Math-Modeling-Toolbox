```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fminbound
```


# Maximización de Beneficio en una Fábrica (Derivadas)

**Problema:**

Una fábrica produce unidades a un costo dado por la función:
$C(x) = 200x + 500$
Donde x es la cantidad producida, $(C(x))$ es el costo total de producir $(x)$ unidades.
El precio de venta está dado por la función:
$P(x) = 400 - 5x$
Donde $(P(x))$ es el precio al que se venden las $(x)$ unidades.
Preguntas:
1. Encuentra la cantidad de producción que
maximiza el beneficio.
2. ¿Cuál es el beneficio máximo?

# Desarrollo matemático

1. **Función de ingreso total**  
   El ingreso total \(I(x)\) es el producto de la cantidad producida (\(x\)) y el precio por unidad (\(P(x)\)):  
   $$
   I(x) = x \cdot P(x) = x \cdot (400 - 5x) = -5x^2 + 400x
   $$

2. **Función de beneficio**  
   El beneficio \(B(x)\) se obtiene restando el costo total \(C(x)\) del ingreso total \(I(x)\):  
   $$
   B(x) = I(x) - C(x)
   $$  
   Sustituyendo las expresiones de \(I(x)\) y \(C(x)\):  
   $$
   B(x) = (400x - 5x^2) - (200x + 500) = -5x^2 + 200x - 500
   $$

3. **Se calcula la derivada de la función beneficio**  
$$
\frac{dB}{dx} = -10x + 200
$$

4. **Se quiere conocer la cantidad de producción que maximia el beneficio, es decir halllar $x$, por lo cual se iguala la derivada a 0, así:**
$$
-10x + 200 = 0 \\
-10x = -200\\
x = 20
$$
5. **Calcular el beneficio maximo**
Se sustituye $x=20$ en la función benificio
$$ B(20) = -5(20)^2 + 200(20) - 500 \\
B(20) = 1500
$$
Así, el beneficio maximo es **1500 unidades monetarias**









```python


#definiendo funciones

def costo(x):
    return 200*x + 500

def precio(x):
    return 400 - 5*x

def ingreso(x):
    return x*precio(x)

def beneficio(x):
    return ingreso(x) - costo(x)

x_vals = np.linspace(0, 50, 500)
beneficio_vals = beneficio(x_vals)

x_max_beneficio = fminbound(lambda x: -beneficio(x), 0, 50)

# Calcular el beneficio máximo
beneficio_maximo = beneficio(x_max_beneficio)

# Imprimir resultados
print(f'El valor de x que maximiza el beneficio es: {x_max_beneficio}')
print(f'El beneficio máximo es: {beneficio_maximo}')


# Graficar la función de beneficio
plt.figure(figsize=(10, 6))  # Tamaño de la figura
plt.plot(x_vals, beneficio_vals, label='Función Beneficio', color='blue')
plt.axvline(x=x_max_beneficio, color='red', linestyle='--', label='Máximo Beneficio')  # Línea discontinua
plt.scatter(x_max_beneficio, beneficio_maximo, color='red', label='Punto Máximo')  # Punto máximo
plt.title('Función de Beneficio')  # Título de la gráfica
plt.xlabel('Cantidad producida (x)')  # Etiqueta del eje x
plt.ylabel('Beneficio')  # Etiqueta del eje y
plt.legend()  # Leyenda
plt.grid()  # Rejilla
plt.show()  # Mostrar gráfica
```

# Ejercicio 2

A manufacturer estimates that when **q** thousand units of a particular commodity are produced each month, the total cost will be $C(q) = 0.4q2 + 3q + 40$ thousand dollars, and all q units can be sold at a price of $p(q) = 22.2 − 1.2q$ dollars per unit.


* Determine the level of production that results in maximum profit. What is the maximum profit?

* At what level of production is the average cost per unit $A(q)=\frac{C(q)}{q}$
 minimized?
*  At what level of production is the average cost equal to the marginal cost $C'(q)?$


*Extraido de Calculus for Business Economics & the Social & Life Sciences, Brief 10th Edition*


# Desarrollo Matemático primer item


1.   Función **ingreso total** *(Revenue)* está dada por $ R(q)=q p(q)$
Al reemplazar obtenemos que $$R(q)=q(22.2-1.2q) = 22.2q-1.2q^2=-1.2q^2+22.2q$$


2.   El beneficio o **profit** está dado por $P(q)=R(q)-C(q)$, así:
$$P(q)=-1.2q^2+22.2q -(0.4q^2+3q+40)$$
$$P(q) = -1.2q^2+22.2q-0.4q^2-3q-40$$
$$ P(q)=−1.6q^2+19.2q-40$$

3. Queremos conocer el **nivel de produccion que permite el **maximo beneficio**. Sabemos que **q** representa este nivel de producción, por lo cual se derivará la función beneficio ( o profit) y se hallara el valor de **q**, así:
$$
\frac{dP}{dq} = −3.2q+19.2
$$
Igualando a 0 para hallar **q**, se tiene que:
$$
\frac{dP}{dq}=0
$$

$$
−3.2q+19.2=0
$$
$$
−3.2q=−19.2$$
$$q= \frac{-19.2}{-3.2}=6$$
Al reemplazar en la función **profit**, se tiene que
$$ P(6)= -1.6(6)^2+19.2(6)-40=17.6$$





```python
def costo(q):
    return 0.4*q**2 + 3*q +40

def precio(q):
    return 22.2 - 1.2*q

def ingreso(q):
    return q*precio(q)

def beneficio(q):
    return ingreso(q) - costo(q)

q_vals = np.linspace(0, 10, 500)

beneficio_vals = beneficio(q_vals)

q_max_beneficio = fminbound(lambda q: -beneficio(q), 0, 10)
beneficio_maximo = beneficio(q_max_beneficio)

# Mostrar resultados
print(f"La cantidad de producción que maximiza el beneficio es: {q_max_beneficio:.2f} unidades")
print(f"El beneficio máximo es: {beneficio_maximo:.2f} unidades monetarias")

# Graficar la función de beneficio
plt.figure(figsize=(10, 6))
plt.plot(q_vals, beneficio_vals, label="Beneficio", color="blue")
plt.axvline(x=q_max_beneficio, color="red", linestyle="--", label=f"Máximo beneficio en x = {q_max_beneficio:.2f}")
plt.title("Evolución del Beneficio según la Cantidad de Producción")
plt.xlabel("Cantidad producida (x)")
plt.ylabel("Beneficio")
plt.grid(True)
plt.legend()
plt.show()
```
